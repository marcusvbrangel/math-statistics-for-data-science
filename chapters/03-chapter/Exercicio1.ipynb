### Exercício 1: Cálculo de Área de um Círculo

Escreva um programa que calcula a área de um círculo. O usuário deve fornecer o raio e o programa deve retornar a área.

```python
import math

def calcular_area_circulo(raio):
    if raio < 0:
        raise ValueError("O raio nao pode ser negativo.")
    return math.pi * raio ** 2

try:

    raio = float(input("Entre com o tamanho do raio: "))

    area_circulo = calcular_area_circulo(raio)

    print(f"A área do circulo é: {area_circulo:.2f}")

except ValueError as e:
    print(f"Erro: Entrada inválida. {e}")    
```

### Exercício 2: Conversão de Temperatura

Crie uma função que converte a temperatura de Celsius para Fahrenheit e vice-versa. O usuário deve informar a temperatura e a escala de origem.

```python
def converter_celsius_para_fahrenheit(c):
    return (c * 1.8) + 32

def converter_fahrenheit_para_celsius(f):
    return (f - 32) / 1.8


resultado_fahrenheit = converter_celsius_para_fahrenheit(100)

print(f"100 graus de celsius para fahrenheit é: {resultado_fahrenheit:.2f}")


resultado_celsius = converter_fahrenheit_para_celsius(212)

print(f"212 graus de fahrenheit para celsius é: {resultado_celsius:.2f}")
```

### Exercício 3: Resolver uma Equação Linear

Faça um programa que resolve uma equação linear do tipo ax + b = 0. Os valores de a e b são fornecidos pelo usuário.

**Nota:**

Uma Equação Linear é uma igualdade matemática onde o maior expoente das incógnitas (letras) é igual a 1.

- Elas não possuem termos ao quadrado, ao cubo ou letras multiplicadas entre si.

- O formato padrão mais comum com duas variáveis é \(y = ax + b\).

- Nela, \(x\) e \(y\) são as variáveis, enquanto \(a\) e \(b\) são números fixos.

- Geometricamente, quando representada em um gráfico cartográfico, o seu resultado gera sempre uma linha reta.

- O termo \(a\) (coeficiente angular) define a inclinação e direção da reta no gráfico.

- O termo \(b\) (coeficiente linear) indica o ponto exato onde a reta corta o eixo vertical Y.

- São usadas para prever tendências simples, como o custo de uma corrida de táxi com taxa fixa.

```python
# O 'x' deve ser um parâmetro que muda, 'a' e 'b' são os coeficientes fixos da reta
def calcular_y_linear(x, a, b):
    return a * x + b

# Definindo uma reta fixa (ex: y = 2x + 1)
coeficiente_a = 2
coeficiente_b = 1

# Testando para um x específico (x = 3)
x_teste = 1
y_resultado = calcular_y_linear(x_teste, coeficiente_a, coeficiente_b)
print(f"Resultado: para x = {x_teste}, o valor de y é = {y_resultado}")

#-----------------
# Gerando pontos corretos para plotar uma linha reta no gráfico
lista_x = [0, 1, 2, 3, 4, 5]

# Passamos cada 'item_x' como o X da fórmula, mantendo 'a' e 'b' fixos
lista_y = [calcular_y_linear(item_x, coeficiente_a, coeficiente_b) for item_x in lista_x]

print("Valores de X:", lista_x)
print("Valores de Y:", lista_y)
```

```python
import matplotlib.pyplot as plt

# plotar um grafico

# --- BLOCO GRÁFICO ---
# 1. Cria a linha ligando os pontos (X, Y)
plt.plot(lista_x, lista_y, label="y = 2x + 1", color="blue", marker="o")

# 2. Configura os títulos e identificações dos eixos
plt.title("Gráfico de uma Equação Linear")
plt.xlabel("Eixo X (Variável Independente)")
plt.ylabel("Eixo Y (Resultado)")

# 3. Adiciona linhas de grade e a legenda da reta
plt.grid(True)
plt.legend()

# 4. Abre a janela para exibir o gráfico na tela
plt.show()
```

### Exercício 4: Cálculo de Média Aritmética

Escreva um programa que calcula a média aritmética de uma lista de números fornecida pelo usuário.

```python
def calcular_media_aritmetica(valores):
    valores_float = [float(valor) for valor in valores]
    return sum(valores_float) / len(valores_float)

lista_valores = "1 2 3 4 5".split()

resultado = calcular_media_aritmetica(lista_valores)

print(f"A média aritmética é: {resultado:.2f}")
```

### Exercício 5: Determinar as Raízes de uma Equação Quadrática (Equação do 2º grau)

Crie um programa que encontra as raízes de uma equação quadrática do tipo ax^2 + bx + c = 0. O programa deve lidar com todos os tipos de raízes (reais e complexas).

**Nota:**

Uma Equação Quadrática (ou de 2º grau) é uma igualdade matemática com formato padrão \(ax^2 + bx + c = 0\).

- Nela, \(a\), \(b\) e \(c\) são números conhecidos, sendo obrigatório que o \(a\) seja diferente de zero.

- O maior expoente da incógnita \(x\) é 2, o que define o comportamento dinâmico dessa função.

- Geometricamente, em vez de uma linha reta, o seu gráfico gera uma curva simétrica chamada parábola.

- Se o termo \(a\) for positivo, a curva aponta para cima (sorriso); se for negativo, aponta para baixo.

- As raízes (ou zeros) da equação são os valores de \(x\) que tornam a igualdade verdadeira, ou seja, onde \(y = 0\).

- No gráfico da parábola, as raízes representam os pontos exatos onde a curva cruza o eixo horizontal X.

- Uma equação quadrática pode ter duas raízes reais, apenas uma, ou nenhuma raiz real.

- Quem define essa quantidade é o Delta (\(\Delta = b^2 - 4ac\)): positivo dá duas, zero dá uma, e negativo dá nenhuma.

- O método mais famoso do mundo para encontrar essas raízes é a conhecida Fórmula de Bhaskara.

- Elas são usadas na vida real para calcular a trajetória de projéteis, pontes suspensas e lucros máximos.

```python
import cmath # biblioteca para trabalhar com numeros complexos

def resolver_equacao_quadratica(a, b, c):
    
    if a == 0:
        return "Erro: O coeficiente 'a' não pode ser zero em uma equação quadrática"
    
    # 1) calcula o valor de delta
    delta = b ** 2 - 4 * a * c
    print(f"\n O valor de delta é: {delta}")

    # 2) analisa o delta para encontrar as raizes
    if (delta > 0):
        # duas raizes reais e diferentes
        x1 = (-b + (delta ** 0.5)) / (2 * a)
        x2 = (-b - (delta ** 0.5)) / (2 * a)
        return f"Possui duas raízes reais:\n x1 = {x1:.2f}\n x2 = {x2:.2f}"
    
    elif (delta == 0):
        # uma unica raiz real (ou duas raizes iguais)
        x = -b / (2 * a)
        return f"Possui uma única raiz real:\ nx = {x:.2f}"
    
    else:
        # delta negativo: duas raizes complexas/imaginarias
        raiz_delta_complexa = cmath.sqrt(delta)
        x1 = (-b + raiz_delta_complexa) / (2 * a)
        x2 = (-b - raiz_delta_complexa) / (2 * a)
        return f"Não possui raízes reais. As raízes complexas são:\ nx1 = {x1}\n x2 = {x2}"

# --- Execução do Programa ---
print("=== CALCULADORA DE EQUAÇÃO QUADRÁTICA ===")

a = 1 
b = -5 
c = 6

resultado = resolver_equacao_quadratica(a, b, c)

print(resultado)
```

### Exercício 6: Calcular o Fatorial de um Número

Escreva uma função que calcula o fatorial de um número inteiro fornecido pelo usuário.

**Nota:**

O Fatorial (\(n!\)) é a multiplicação de um número inteiro positivo por todos os seus antecessores até o 1.

- O cálculo segue a estrutura decrescente: por exemplo, \(4! = 4 . 3 . 2 . 1 = 24\).

- Por convenção e definição matemática, o fatorial do número zero (\(0!\)) é sempre igual a 1.

- Seu crescimento é extremamente veloz, aumentando muito mais rápido que potências ou juros compostos.

**Para que serve:**

- Ele é a base da Análise Combinatória, usada para contar possibilidades de agrupamento.

- Serve para calcular permutações, descobrindo quantas formas diferentes existem para organizar uma fila.

- Serve para calcular combinações, determinando quantos jogos possíveis existem em sorteios como a Mega-Sena.

- É aplicado na computação para analisar a complexidade de algoritmos e testar caminhos de rede.

- É usado na estatística e probabilidade para prever chances de eventos acontecerem em jogos ou física.

```python
def calcular_fatorial(numero):
    # Valida se o número é negativo
    if numero < 0:
        return "Não existe fatorial para números negativos."
    
    fatorial = 1
    # Multiplica os números do número fornecido regressivamente até 1
    # 4! = 4 . 3 . 2 . 1 = 24
    for i in range(numero, 0, -1):
        fatorial *= i
                
    return fatorial

# Executa o teste com o número 4
try:
    numero = 4
    resultado = calcular_fatorial(numero)
    print(f"O fatorial de {numero} é: {resultado}")
except ValueError:
    print("Por favor, digite apenas números inteiros válidos.")
```

### Exercício 7: Sequência de Fibonacci

Crie um programa que gera a sequência de Fibonacci até um determinado número n fornecido pelo usuário.

**Note:**

A Sequência de Fibonacci é uma sucessão infinita de números inteiros que começa por 0 e 1.

- Cada termo seguinte é a soma dos dois números anteriores (ex: 0, 1, 1, 2, 3, 5, 8, 13...).

- Ela foi criada pelo matemático Leonardo Pisa para descrever o crescimento de uma população de coelhos.

- Na natureza, essa lógica explica o padrão de crescimento de girassóis, conchas e galhos de árvores.

```python
def gerar_fibonacci(limite):
    # Começamos com os dois primeiros números da sequência
    a = 0
    b = 1
    
    print(f"Sequência de Fibonacci até {limite}:")
    
    # O loop continua enquanto o número atual for menor ou igual ao limite
    while a <= limite:
        print(a, end=" ") # Mostra o número na mesma linha
        
        # A mágica acontece aqui: 
        # 'a' vira o próximo número (b)
        # 'b' vira a soma dos dois anteriores (a + b)
        a, b = b, a + b
    print() # Pula uma linha no final

# Teste do programa com o limite de 50
gerar_fibonacci(28)
```

### Exercício 8: Cálculo de Juros Compostos

Faça um programa que calcula o montante final de um investimento com juros compostos. O usuário deve fornecer o capital inicial, a taxa de juros anual, e o número de anos.

**Nota:**

Os Juros Compostos representam o conceito de "juros sobre juros", muito comum no sistema financeiro mundial.

- Ao contrário dos juros simples, a taxa incide sobre o valor acumulado e atualizado do período anterior.

- Isso significa que o rendimento ou a dívida cresce de forma exponencial, e não linear.

- O montante acumulado aumenta mais rápido a cada mês que passa, acelerando como uma bola de neve.

- No início o crescimento parece lento, mas a longo prazo o efeito dos juros sobre juros é massivo.

- A fórmula matemática usada para calcular o valor final é expressa por \(M = P \cdot (1 + i)^t\).

- Nela, \(M\) é o montante, \(P\) o valor inicial, \(i\) a taxa de juros e \(t\) o tempo de aplicação.

- São a base dos rendimentos de investimentos e também do crescimento das dívidas do cartão de crédito.

```python
def calcular_juros_compostos(capital, taxa_anual, anos):

    # converte a taxa em porcentagem
    taxa_decimal = taxa_anual / 100

    # Aplica a fórmula matemática: M = P * (1 + i)^t
    montante = capital * (1 + taxa_decimal) ** anos

    return montante


# investimento de R$ 1.000,00 a uma taxa de 10% ao ano durante 3 anos
try:
    # Coleta de dados do usuário
    capital_inicial = 1000
    taxa_juros_anual = 10
    tempo_anos = 3

    # Executa o cálculo
    montante_final = calcular_juros_compostos(capital_inicial, taxa_juros_anual, tempo_anos)
    lucro = montante_final - capital_inicial

    # Exibe os resultados na tela com duas casas decimais
    print("\n=== RESULTADO DA SIMULAÇÃO ===")
    print(f"Total acumulado (Montante): R$ {montante_final:.2f}")
    print(f"Total ganho em juros (Lucro): R$ {lucro:.2f}")

except ValueError:
    print("Erro: Por favor, insira valores numéricos válidos.")
```

### Exercício 10: Verificar se um Número é Primo

Crie uma função que verifica se um número fornecido pelo usuário é primo ou não.

**Nota:**

Um número primo é um número natural maior do que 1 que possui exatamente dois divisores positivos distintos: 

- O número 1 e ele mesmo. 

- Se um número puder ser dividido por outros algarismos além desses, ele é chamado de número composto. 

- O número 2 é o único primo que também é par, e todos os demais números primos são ímpares. 

- Exemplos de números primos incluem 2, 3, 5, 7, 11 e 13.

```python
def verificar_primo(numero):

    if (numero <= 1):
        return  # numero menor ou igual a 1 nao e primo

    # verifica se o numero tem outros divisores
    # testar divisores do numero no loop até a raiz quadrada do número.
    # Se o código encontra um divisor, ele imprime que não é primo e para a execução ali mesmo.
    # numero**0.5: Calcula a raiz quadrada do número.

    # Se um número tem um divisor maior que a sua raiz quadrada, 
    # ele obrigatoriamente terá outro par divisor menor do que ela. 
    # Se você testar até a raiz quadrada e não encontrar nenhum divisor, 
    # o número com certeza é primo
    print(f"numero {numero}: raiz do número: {numero ** 0.5}, raiz do numero convertido/arreondado para int + 1: {int(numero ** 0.5) + 1} ")
    for i in range(2, int(numero ** 0.5) + 1):
        if numero % i == 0:
            print(f"Número {numero} não é primo")
            return
        
    print(f"Número {numero} é primo")


# testando a lista de 1 ate 16
lista_numeros = list(range(1, 16))

_ = [verificar_primo(numero) for numero in lista_numeros]
```
