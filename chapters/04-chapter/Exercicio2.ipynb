### Exercício 1: Gráfico de Uma Função Linear

Crie um gráfico para a função y = 2x + 3 para valores de x no intervalo de 5 a 10.

**Nota:**

Uma função linear (Primeiro Grau) é uma regra matemática que relaciona duas variáveis com um crescimento constante. 

Sua fórmula geral é \(y = ax\), onde \(a\) é um número fixo diferente de zero. 

O gráfico dessa função é sempre uma linha reta que obrigatoriamente passa pela origem \((0,0)\). 

Se a variável \(x\) dobra de valor, a variável \(y\) também dobrará proporcionalmente

```python
def calcular_funcao_linear(x):
    return 2*x + 3

valores_x = list(range(5, 11))
valores_y = [calcular_funcao_linear(valor_x) for valor_x in valores_x]

print(valores_x)
print(valores_y)

plt.figure(figsize=(8, 4))  # largura e altura em polegadas
plt.plot(valores_x, valores_y, label="y = 2x + 3", color="green", marker="o")
plt.title("Gráfico de uma Função Linear")
plt.xlabel("Eixo X (Variável Independente)")
plt.ylabel("Eixo Y Resultado(Variável Dependente)")
plt.legend()
plt.grid(True)
plt.show()
```

### Exercício 2: Gráfico de uma Função Quadrática

Desenhe o gráfico da função y = x^2 - 4x + 4. Crie uma massa de dados de exemplo usando np.linspace().

**Nota:**

Uma função quadrática (Segundo Grau) é uma regra matemática onde a variável de maior grau está elevada ao quadrado. 

Sua fórmula geral é \(y = ax^2 + bx + c\), sendo \(a\) obrigatoriamente diferente de zero. 

O gráfico dessa função tem o formato de uma curva chamada parábola. 

Se o valor de \(a\) for positivo, a curva se abre para cima; se for negativo, para baixo. 

Ela possui um ponto máximo ou mínimo chamado vértice e é muito usada para calcular trajetórias e áreas

```python
def calcular_funcao_quatratica(x):
    return x**2 -4*x + 4

valores_x = np.linspace(-6, 10, 200)
valores_y = [calcular_funcao_quatratica(valor_x) for valor_x in valores_x]

# print(valores_x)
# print(valores_y)

plt.figure(figsize=(8, 6)) # largura e altura em polegadas
plt.plot(valores_x, valores_y, label="y = ax^2 + bx + c", color="green", marker="x")
plt.title("Gráfico de uma Função Quadrática")
plt.xlabel("Eixo X (Variável Independente)")
plt.ylabel("Eixo Y (Variável Dependente)")
plt.legend()
plt.grid(True)
plt.show()
```
