### Exercício 1: Gráfico de Uma Função Linear

Crie um gráfico para a função y = 2x + 3 para valores de x no intervalo de 5 a 10.

**Nota:**

Uma função linear (Primeiro Grau) é uma regra matemática que relaciona duas variáveis com um crescimento constante. 

Sua fórmula geral é \(y = ax\), onde \(a\) é um número fixo diferente de zero. 

O gráfico dessa função é sempre uma linha reta que obrigatoriamente passa pela origem \((0,0)\). 

Se a variável \(x\) dobra de valor, a variável \(y\) também dobrará proporcionalmente

```python
def calcular_funcao_linear(x):
    return 2*x + 3

valores_x = list(range(5, 11))
valores_y = [calcular_funcao_linear(valor_x) for valor_x in valores_x]

print(valores_x)
print(valores_y)

plt.figure(figsize=(8, 4))  # largura e altura em polegadas
plt.plot(valores_x, valores_y, label="y = 2x + 3", color="green", marker="o")
plt.title("Gráfico de uma Função Linear")
plt.xlabel("Eixo X (Variável Independente)")
plt.ylabel("Eixo Y Resultado(Variável Dependente)")
plt.legend()
plt.grid(True)
plt.show()
```

### Exercício 2: Gráfico de uma Função Quadrática

Desenhe o gráfico da função y = x^2 - 4x + 4. Crie uma massa de dados de exemplo usando np.linspace().

**Nota:**

Uma função quadrática (Segundo Grau) é uma regra matemática onde a variável de maior grau está elevada ao quadrado. 

Sua fórmula geral é \(y = ax^2 + bx + c\), sendo \(a\) obrigatoriamente diferente de zero. 

O gráfico dessa função tem o formato de uma curva chamada parábola. 

Se o valor de \(a\) for positivo, a curva se abre para cima; se for negativo, para baixo. 

Ela possui um ponto máximo ou mínimo chamado vértice e é muito usada para calcular trajetórias e áreas

```python
def calcular_funcao_quatratica(x):
    return x**2 -4*x + 4

valores_x = np.linspace(-6, 10, 200)
valores_y = [calcular_funcao_quatratica(valor_x) for valor_x in valores_x]

# print(valores_x)
# print(valores_y)

plt.figure(figsize=(8, 6)) # largura e altura em polegadas
plt.plot(valores_x, valores_y, label="y = ax^2 + bx + c", color="green", marker="x")
plt.title("Gráfico de uma Função Quadrática")
plt.xlabel("Eixo X (Variável Independente)")
plt.ylabel("Eixo Y (Variável Dependente)")
plt.legend()
plt.grid(True)
plt.show()
```

### Exercício 3: Gráfico de Uma Função Senoidal

Crie um gráfico da função y = sin(x) para valores de x no intervalo de 0 a 2π.

**Nota:**

Uma função senoidal é um modelo matemático que descreve fenômenos periódicos ou oscilatórios através da função seno. 

Sua fórmula básica é y = sen(x), gerando um gráfico em formato de onda suave que se repete infinitamente, chamado de senoide. 

Ela varia harmonicamente entre um valor máximo e um mínimo fixos, passando pelo zero de forma regular. 

É fundamental para estudar movimentos de vai e vem, como ondas de som, luz, marés e corrente elétrica alternada.

```python
def calcular_funcao_senoidal(x):
    return np.sin(x)

valores_x = np.linspace(0, 2 * np.pi, 200)
valores_y = [calcular_funcao_senoidal(valor_x) for valor_x in valores_x]

plt.figure(figsize=(8, 6)) # largura e altura em polegadas
plt.plot(valores_x, valores_y, label="y = sin(x)", color="green", marker="x")
plt.title("Gráfico de uma Função Senoidal")
plt.xlabel("Eixo X (Variável Independente)")
plt.ylabel("Eixo Y (Variável Dependente)")
plt.legend()
plt.grid(True)
plt.show()
```

### Exercício 4: Gráfico de Uma Função Exponencial

Crie um gráfico para a função y = e^x para valores de x no intervalo de -2 a 2.

**Nota:**

Uma função exponencial é uma regra matemática onde a variável independente fica no expoente. 

Sua fórmula geral é (y = a^x), onde a base (a) deve ser positiva e diferente de um. 

O gráfico dessa função é uma curva que cresce ou decresce de forma extremamente rápida. 

Quando a base é maior que um, o crescimento acelera a cada passo; quando está entre zero e um, o valor reduz drasticamente. 

Ela é muito usada para modelar o crescimento de bactérias, juros compostos e a propagação de vírus.

```python
def calcular_funcao_exponencial(x):
    return 2**x

valores_x = np.linspace(0.1, 10, 100)
valores_y = [calcular_funcao_exponencial(valor_x) for valor_x in valores_x]

plt.figure(figsize=(8, 6)) # largura e altura em polegadas
plt.plot(valores_x, valores_y, label="y = a^x", color="green", marker="x")
plt.title("Gráfico de uma Função Exponencial")
plt.xlabel("Eixo X (Variável Independente)")
plt.ylabel("Eixo Y (Variável Dependente)")
plt.legend()
plt.grid(True)
plt.show()
```

### Exercício 5: Gráfico de uma Função Logarítmica: 

Desenhe o gráfico de y = log(x) (logaritmo natural) para valores de x no intervalo de 0.1 a 10.

**Nota:**

Uma função logarítmica é o inverso direto da função exponencial, usada para descobrir o expoente necessário para obter determinado número. 

Sua fórmula geral é (y = log_a(x)), onde a base (a) deve ser positiva e diferente de um, e (x) deve ser maior que zero. 

O gráfico dessa função é uma curva que cresce de forma muito lenta à medida que (x) aumenta. 

Ela sempre cruza o eixo horizontal no ponto ((1,0)) e nunca toca o eixo vertical (Y). 

É a ferramenta matemática ideal para medir grandezas que variam em escalas gigantescas. 

Por isso, ela é muito aplicada na prática para calcular o pH de soluções, a escala Richter de terremotos e a intensidade do som em decibéis.

```python
def calcular_funcao_logaritmica(x):
    return np.log(x)

valores_x = np.linspace(0.1, 10, 100)
valores_y = [calcular_funcao_logaritmica(valor_x) for valor_x in valores_x]
 
plt.figure(figsize=(8, 6)) # largura e altura em polegadas
plt.plot(valores_x, valores_y, label="y = log_a(x)", color="green", marker="x")
plt.title("Gráfico de uma Função Logaritmica")
plt.xlabel("Eixo X (Variável Independente)")
plt.ylabel("Eixo Y (Variável Dependente)")
plt.legend()
plt.grid(True)
plt.show()
```

### Exercício 6: Resolver uma Equação Quadrática

Escreva um script em Python que resolve a equação x^2 − 5x + 6 = 0 e imprime as raízes.

**Nota:**

Uma equação quadrática, ou equação do segundo grau, é uma igualdade matemática onde o maior expoente da incógnita é igual a dois. 

Sua forma geral padrão é descrita como (ax^2 + bx + c = 0), na qual (a), (b) e (c) são coeficientes reais e (a) deve ser necessariamente diferente de zero. 

Ao contrário da função, que mapeia uma relação entre duas variáveis reais, a equação busca encontrar valores específicos para uma única incógnita que tornem a igualdade verdadeira. 

Esses valores numéricos são chamados de raízes ou soluções da equação, existindo no máximo duas respostas reais possíveis. 

O método de resolução mais famoso envolve o cálculo do discriminante Delta ((Delta = b^2 - 4ac)) seguido pela aplicação da fórmula de Bhaskara. 

Se o Delta for positivo, a equação terá duas raízes reais distintas; se for zero, terá duas raízes reais iguais; e se for negativo, não possuirá raízes reais. 

Ela é amplamente aplicada na física para determinar o exato instante de tempo em que um objeto em movimento atinge o solo

```python
# passo 01 - definir os coeficientes da equacao (x² - 5x + 6 = 0)
a = 1
b = -5
c = 6

# calculo do delta
delta = b**2 - 4 * a * c

# verificacao do delta e calculo da raizes
if delta > 0:
    raiz_delta = math.sqrt(delta)
    x1 = (-b + raiz_delta) / (2 * a)
    x2 = (-b - raiz_delta) / (2 * a)
    print(f"A equação possui duas raízes reais distintas:")
    print(f"x1 = {x1}")
    print(f"x2 = {x2}")
elif delta == 0:
    x = -b (20 * a)    
    print(f"A equação possui duas raízes reais iguais:")
    print(f"x1 = x2 = {x}")
else:
    print("A equação não possui raízes reais")    
```

### Exercício 8: Equação Quadrática com Discriminante

Para a equação quadrática ax^2 + bx + c = 0, calcule o discriminante e determine o número de raízes reais.

**Nota:**

Uma equação quadrática com discriminante refere-se ao estudo e resolução de uma equação do segundo grau (ax^2 + bx + c = 0) focando na análise do elemento conhecido como Delta (Delta). 

O discriminante é uma expressão algébrica específica, calculada pela fórmula (Delta = b^2 - 4ac), utilizando os coeficientes numéricos da própria equação. 

A palavra "discriminante" vem de "discriminar", que significa distinguir ou diferenciar, pois o valor desse número revela a natureza das raízes antes de calculá-las. 

Ele serve como um indicador do comportamento matemático e geométrico da equação no conjunto dos números reais.

Existem três cenários possíveis determinados pelo valor do discriminante:

- **Delta maior que zero (Delta > 0):** A equação possui duas raízes reais e totalmente diferentes. No gráfico, a parábola associada corta o eixo horizontal X em dois pontos distintos.

- **Delta igual a zero (Delta = 0):** A equação possui duas raízes reais idênticas (uma raiz única). Geometricamente, isso significa que o vértice da parábola apenas toca (tangencia) o eixo (X) em um único ponto.

- **Delta menor que zero (Delta < 0):** A equação não possui nenhuma raiz real, pois não existe raiz quadrada de número negativo nesse conjunto. O gráfico flutua totalmente acima ou abaixo do eixo (X).

Essa análise é uma ferramenta essencial na álgebra para prever o comportamento de sistemas físicos, trajetórias e lucros de forma rápida e eficiente.

```python
# passo 01 - definicao dos coeficientes

# Teste 1: Duas Raízes Reais (Delta Positivo)
# Resultado esperado: Delta = 1.00 (A equação possui 2 raízes reais e diferentes).
a = 1
b = -5
c = 6
 
# Teste 2: Uma Única Raiz Real (Delta Igual a Zero)
# Resultado esperado: Delta = 0.00 (A equação possui 1 única raiz real).
# a = 1
# b = -4
# c = 4

# Teste 3: Nenhuma Raiz Real (Delta Negativo)
# Resultado esperado: Delta = -16.00 (A equação possui 0 raízes reais).
# a = 1
# b = 2
# c = 5

# passo 02 - Cálculo do discriminante (Delta)
delta = b**2 - 4 * a * c

# passo 03 - Exibição do valor do Delta
print(f"\nO valor do discriminante (Delta) é: {delta:.2f}")

# passo 04 -  Determinação do número de raízes reais
if delta > 0:
    print("A equação possui 2 raízes reais e diferentes.")
    print(f"x1 = {x1:.2f}")
    print(f"x2 = {x2:.2f}")
elif delta == 0:
    print("A equação possui 1 única raiz real (2 raízes reais e iguais).")
    print(f"x1 = {x1:.2f}")
    print("x2 não existe separadamente (as duas raízes são iguais).")
else:
    print("A equação possui 0 raízes reais (as raízes são complexas).")
```
